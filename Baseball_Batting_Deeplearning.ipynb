```python
!cp /content/drive/MyDrive/best_top1_acc_epoch_14.pth /content/mmaction2/cheakpoints/
!cp /content/drive/MyDrive/tsm_1x1x8_25e_baseball_data_1116.py /content/mmaction2/configs/recognition/tsm/
!cp /content/drive/MyDrive/2022-06-10_03-20-30.mp4 /content/mmaction2/videos/
!cp /content/drive/MyDrive/baseball_label.txt /content/mmaction2/videos/

%cd content/mmaction2

from mmaction.apis import inference_recognizer, init_recognizer

# Choose to use a config and initialize the recognizer
config = 'configs/recognition/tsm/tsm_1x1x8_25e_baseball_data_1116.py'
# Setup a checkpoint file to load
checkpoint = 'cheakpoints/best_top1_acc_epoch_14.pth'
# Initialize the recognizer
model = init_recognizer(config, checkpoint, device='cuda:0')
```

```python
# Use the recognizer to do inference
video = 'videos/2022-06-10_03-20-30.mp4'
label = 'videos/baseball_label.txt'
results = inference_recognizer(model, video)

labels = open(label).readlines()
labels = [x.strip() for x in labels]
results = [(labels[k[0]], k[1]) for k in results]
```

```python
# Let's show the results
for result in results:
    print(f'{result[0]}: ', result[1])
```
